```python
import os
import shutil

files_to_move = [
    "interactions.csv",
    "targets.csv",
    "editions.csv",
    "book_genres.csv",
    "users.csv",
]

for name in files_to_move:
    if os.path.exists(name):
        shutil.move(name, f"data/{name}")

print("done")
```
